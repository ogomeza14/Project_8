```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Source d'information sur les exportations et importations de la France : Institut National de la Statistique et des Études Économiques (INSEE)

web: https://www.insee.fr/fr/statistiques?theme=27

```python
FR_X_prod = pd.read_csv('C:/Users/Oscar Gomez/Desktop/Data Analyst/P8/FR_X_products.csv', sep=';')
FR_X_prod.drop(columns = ['code'], inplace=True)
FR_X_prod.rename(columns = {'product': 'time'}, inplace=True)
FR_X_prod.set_index('time', inplace=True)
FR_X_prod = FR_X_prod.T
FR_X_prod.head()
```

```python
FR_X_prod.index = pd.to_datetime(FR_X_prod.index)
FR_X_prod.head()
```

```python
FR_X_prod['Total'] = FR_X_prod.sum(axis=1)
FR_X_prod.head()
```

```python
FR_X_prod.describe()
```

```python
FR_M_prod = pd.read_csv('C:/Users/Oscar Gomez/Desktop/Data Analyst/P8/FR_M_products.csv', sep=';')
FR_M_prod.drop(columns = ['code'], inplace=True)
FR_M_prod.rename(columns = {'product': 'time'}, inplace=True)
FR_M_prod.set_index('time', inplace=True)
FR_M_prod = FR_M_prod.T
FR_M_prod.head()
```

```python
FR_M_prod.index = pd.to_datetime(FR_M_prod.index)
FR_M_prod.head()
```

```python
FR_M_prod['Total'] = FR_M_prod.sum(axis=1)
FR_M_prod.head()
```

```python
FR_M_prod.describe()
```

```python
plt.figure(figsize=(15,6))

plt.subplot()
plt.plot(FR_X_prod.index, FR_X_prod['Total'], label='Exportations totales', color='r')
plt.plot(FR_M_prod.index, FR_M_prod['Total'], label='Importations totales', color='g')
plt.ylim(0, 65000)

plt.title('Commerce International de France \n(2000-2021)', fontsize=20)
plt.ylabel('Millions d\'euros', fontsize=16)
plt.xlabel('Années', fontsize=16)
plt.legend(loc='lower right')

plt.savefig('P8_FR_commtotal.png');
```

```python
FR_series = pd.DataFrame(FR_X_prod['Total'])
FR_series.rename(columns={'Total': 'expo'}, inplace=True)
FR_series['impo'] = FR_M_prod['Total']
FR_series['bal_trade'] = FR_series['expo'] - FR_series['impo']
FR_series.head()
```

```python
plt.figure(figsize=(15,6))

plt.plot(FR_series.index, FR_series['bal_trade'], label='Balance')

plt.title('Balance comerciale de France \n(2000-2021)', fontsize=20)
plt.ylabel('Millions d\'euros', fontsize=16)
plt.xlabel('Années', fontsize=16)
plt.legend(loc='upper right')

plt.savefig('P8_FR_balance.png');
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
decomp_FR_expo = seasonal_decompose(FR_series['expo'],  model='additive')
```

```python
obs_FRX = pd.DataFrame(FR_series['expo'])
obs_FRX.reset_index(inplace=True)

trend_FRX = pd.DataFrame(decomp_FR_expo.trend)
trend_FRX.reset_index(inplace=True)

seasonal_FRX = pd.DataFrame(decomp_FR_expo.seasonal)
seasonal_FRX.reset_index(inplace=True)

resid_FRX = pd.DataFrame(decomp_FR_expo.resid)
resid_FRX.reset_index(inplace=True)

df1 = pd.merge(obs_FRX, seasonal_FRX)
df2 = pd.merge(df1, trend_FRX)
FRX_ma = pd.merge(df2, resid_FRX)
FRX_ma.set_index('index', inplace=True)

FRX_ma.tail(15)
```

```python
plt.figure(figsize=(12,10))
plt.subplots_adjust(hspace=0.2)

plt.subplot(411)
plt.plot(FRX_ma.index, FRX_ma['expo'])
plt.ylabel('Exportations', fontsize=12)
plt.axvline(pd.to_datetime('01/01/2000'), linestyle='--', color='grey')
plt.axvline(pd.to_datetime('12/01/2021'), linestyle='--', color='grey')

plt.subplot(412)
plt.plot(FRX_ma.index, FRX_ma['seasonal'], 'r')
plt.ylabel('Saisonnalité', fontsize=12)
plt.axvline(pd.to_datetime('01/01/2000'), linestyle='--', color='grey')
plt.axvline(pd.to_datetime('12/01/2021'), linestyle='--', color='grey')

plt.subplot(413)
plt.plot(FRX_ma.index, FRX_ma['trend'], 'g')
plt.ylabel('Tendance', fontsize=12)
plt.axvline(pd.to_datetime('01/01/2000'), linestyle='--', color='grey')
plt.axvline(pd.to_datetime('12/01/2021'), linestyle='--', color='grey')

plt.subplot(414)
plt.plot(FRX_ma.index, FRX_ma['resid'], 'y')
plt.ylabel('Residus', fontsize=12)
plt.xlabel('Années', fontsize=12)
plt.axvline(pd.to_datetime('01/01/2000'), linestyle='--', color='grey')
plt.axvline(pd.to_datetime('12/01/2021'), linestyle='--', color='grey')

plt.savefig('P8_FRX_season')
plt.show()
```

```python
FR_series['expo_ma'] = FR_series['expo'] - FRX_ma['seasonal']
FR_series.head()
```

```python
decomp_FR_impo = seasonal_decompose(FR_series['impo'],  model='additive')
```

```python
obs_FRM = pd.DataFrame(FR_series['impo'])
obs_FRM.reset_index(inplace=True)

trend_FRM = pd.DataFrame(decomp_FR_impo.trend)
trend_FRM.reset_index(inplace=True)

seasonal_FRM = pd.DataFrame(decomp_FR_impo.seasonal)
seasonal_FRM.reset_index(inplace=True)

resid_FRM = pd.DataFrame(decomp_FR_impo.resid)
resid_FRM.reset_index(inplace=True)

df1 = pd.merge(obs_FRM, seasonal_FRM)
df2 = pd.merge(df1, trend_FRM)
FRM_ma = pd.merge(df2, resid_FRM)
FRM_ma.set_index('index', inplace=True)

FRM_ma.tail(15)
```

```python
plt.figure(figsize=(12,10))
plt.subplots_adjust(hspace=0.2)

plt.subplot(411)
plt.plot(FRM_ma.index, FRM_ma['impo'])
plt.ylabel('Importations', fontsize=12)
plt.axvline(pd.to_datetime('01/01/2000'), linestyle='--', color='grey')
plt.axvline(pd.to_datetime('12/01/2021'), linestyle='--', color='grey')

plt.subplot(412)
plt.plot(FRM_ma.index, FRM_ma['seasonal'], 'r')
plt.ylabel('Saisonnalité', fontsize=12)
plt.axvline(pd.to_datetime('01/01/2000'), linestyle='--', color='grey')
plt.axvline(pd.to_datetime('12/01/2021'), linestyle='--', color='grey')

plt.subplot(413)
plt.plot(FRM_ma.index, FRM_ma['trend'], 'g')
plt.ylabel('Tendance', fontsize=12)
plt.axvline(pd.to_datetime('01/01/2000'), linestyle='--', color='grey')
plt.axvline(pd.to_datetime('12/01/2021'), linestyle='--', color='grey')

plt.subplot(414)
plt.plot(FRM_ma.index, FRM_ma['resid'], 'y')
plt.ylabel('Residus', fontsize=12)
plt.xlabel('Années', fontsize=12)
plt.axvline(pd.to_datetime('01/01/2000'), linestyle='--', color='grey')
plt.axvline(pd.to_datetime('12/01/2021'), linestyle='--', color='grey')

plt.savefig('P8_FRM_season')
plt.show()
```

```python
FR_series['impo_ma'] = FR_series['impo'] - FRM_ma['seasonal']
FR_series.head()
```

```python
plt.figure(figsize=(15,6))

plt.subplot()
plt.plot(FR_series.index, FR_series['expo_ma'], label='Exportations totales', color='r')
plt.plot(FR_series.index, FR_series['impo_ma'], label='Importations totales', color='g')
plt.ylim(0, 65000)

plt.title('Commerce International de France \n(2000-2021)', fontsize=20)
plt.ylabel('Millions d\'euros', fontsize=16)
plt.xlabel('Années', fontsize=16)
plt.legend(loc='lower right')

plt.savefig('P8_FR_commtotalcor.png');
```

```python
plt.figure(figsize=(15,6))

plt.subplot()
plt.plot(FR_X_prod.index, FR_X_prod['Total'], label='Exportations totales', linestyle='--', color='r', alpha=0.5)
plt.plot(FR_M_prod.index, FR_M_prod['Total'], label='Importations totales', linestyle='--', color='g', alpha=0.5)
plt.plot(FR_series.index, FR_series['expo_ma'], label='Exportations totales corr.', color='r')
plt.plot(FR_series.index, FR_series['impo_ma'], label='Importations totales corr.', color='g')
plt.ylim(0, 65000)

plt.title('Commerce International de France \n(2000-2021)', fontsize=20)
plt.ylabel('Millions d\'euros', fontsize=16)
plt.xlabel('Années', fontsize=16)
plt.legend(loc='lower right')

plt.savefig('P8_FR_compcor.png');
```
